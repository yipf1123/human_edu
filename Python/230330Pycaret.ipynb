```python
import pycaret
pycaret.__version__
```

```python
from pycaret.datasets import get_data
data = get_data('diabetes')
```

```python
# set up
from pycaret.classification import *
s = setup(data, target = 'Class variable', session_id = 1)
```

- Stratified k-fold cross-validation은 각각의 학습셋과 검증셋에서 클래스 분포를 유지하도록 분할하는 것입니다. 

- 즉, 각각의 부분 집합에서 클래스 비율이 전체 데이터셋에서의 클래스 비율과 동일하도록 데이터를 샘플링합니다. 이를 통해 모델이 학습할 때, 불균형한 클래스 비율에 영향을 받지 않고 일반화된 모델을 만들 수 있습니다.

```python
best = compare_models()
```
