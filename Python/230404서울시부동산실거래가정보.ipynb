```python
import lxml
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import math


SERVICE_KEY = '564b7852646a686a34336f4f6c5571'
CNTRCT_DE = '20230302'
ACC_YEAR = '2023'
GU = '11500'

url = f'http://openapi.seoul.go.kr:8088/{SERVICE_KEY}/json/tbLnOpendataRtmsV/1/5/2022'

result = requests.get(url)
content = result.content
soup = BeautifulSoup(content, "lxml") # html.parser 
# print(soup)
print(soup.prettify())
```

### 방법 1 - zip 활용
- 반복문을 작성하도록 한다. 

```python
year_list           = []
sgg_cd_list         = []
bldg_nm_list        = []
obj_amt_list        = []
house_type_list     = []
rdealer_lawdnm_list = []

start_time = time.time()

for year, sgg_cd, bldg_nm, obj_amt, house_type, rdealer_lawdnm in zip(years, sgg_cds, bldg_nms, obj_amts, house_types, rdealer_lawdnms):
  year_list.append(year.get_text())
  sgg_cd_list.append(sgg_cd.get_text())
  bldg_nm_list.append(bldg_nm.get_text())
  obj_amt_list.append(obj_amt.get_text())
  house_type_list.append(house_type.get_text())
  rdealer_lawdnm_list.append(rdealer_lawdnm.get_text())

df = pd.DataFrame({
    "acc_year": year_list, 
    "sgg_cd": sgg_cd_list, 
    "bldg_nm" : bldg_nm_list,
    "obj_amt": obj_amt_list,
    "house_type" : house_type_list,
    "rdealer_lawdnm": rdealer_lawdnm_list
})

elapsed_time = time.time() - start_time

hours, remainder = divmod(elapsed_time, 3600)
minutes, seconds = divmod(remainder, 60)

print("Execution time: ", elapsed_time)
print("Execution time: {:02d}:{:02d}:{:02d}".format(int(hours), int(minutes), int(seconds)))

df
```

### 방법 2 - 반복문 활용

```python
data = []

start_time = time.time()

for i in range(0,len(years)):
  rows = [years[i].get_text(),
          sgg_cds[i].get_text(), 
          bldg_nms[i].get_text(),
          obj_amts[i].get_text(),
          house_types[i].get_text(),
          rdealer_lawdnms[i].get_text()]
  data.append(rows)

cols = ['acc_year', 'sgg_cd', 'bldg_nm', 'obj_amt', 'house_type', 'rdealer_lawdnm']
df = pd.DataFrame(data, columns = cols)

elapsed_time = time.time() - start_time

hours, remainder = divmod(elapsed_time, 3600)
minutes, seconds = divmod(remainder, 60)

print("Execution time: ", elapsed_time)
print("Execution time: {:02d}:{:02d}:{:02d}".format(int(hours), int(minutes), int(seconds)))

df
```

### 데이터 추출
- 비교적 데이터는 깔끔한 편이다. 
- 간단하게 데이터를 추출할 수 있다. 
- 시간 측정을 해보도록 한다. 

```python
cols = ['acc_year', 'sgg_cd', 'bldg_nm', 'obj_amt', 'house_type', 'rdealer_lawdnm']
cols_upper = [col.upper() for col in cols]
```

```python
contents = content['tbLnOpendataRtmsV']['row']
records = []

start_time = time.time()

for bldg in contents:
  datas = {}
  datas[cols[0]] = bldg[cols_upper[0]]
  datas[cols[1]] = bldg[cols_upper[1]]
  datas[cols[2]] = bldg[cols_upper[2]]
  datas[cols[3]] = bldg[cols_upper[3]]
  datas[cols[4]] = bldg[cols_upper[4]]
  datas[cols[5]] = bldg[cols_upper[5]]
  # datas[cols[6]] = bldg[cols[6]]
  records.append(datas)

df = pd.DataFrame(records)

elapsed_time = time.time() - start_time

hours, remainder = divmod(elapsed_time, 3600)
minutes, seconds = divmod(remainder, 60)

print("Execution time: ", elapsed_time)
print("Execution time: {:02d}:{:02d}:{:02d}".format(int(hours), int(minutes), int(seconds)))

df
```
